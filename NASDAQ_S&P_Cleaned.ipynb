# NASDAQ Financial Ratios

```python
nasdaq_url = 'https://financialmodelingprep.com/api/v3/nasdaq_constituent?apikey=' + api_key
```

```python
nasdaq_list_response = requests.get(nasdaq_url).json()
```

```python
# collect a list of the NASDAQ constituents
nasdaq_companies = []
company_tickers = []

for company in nasdaq_list_response:
    nasdaq_companies.append(company['name'])
    company_tickers.append(company['symbol'])
```

```python
data = {'Company': company_tickers, 'Q1 P/E': ['' for i in range(len(company_tickers))], 
        'Q2 P/E': ['' for i in range(len(company_tickers))], 
        'Q3 P/E': ['' for i in range(len(company_tickers))], 
        'Q1 ROE': ['' for i in range(len(company_tickers))], 
        'Q2 ROE': ['' for i in range(len(company_tickers))], 
        'Q3 ROE': ['' for i in range(len(company_tickers))], 
        'Q1 Date': ['' for i in range(len(company_tickers))],
        'Q2 Date': ['' for i in range(len(company_tickers))],
        'Q3 Date': ['' for i in range(len(company_tickers))],
       }
nasdaq_df = pd.DataFrame(data=data)
nasdaq_df.head()
```

```python
# gather data on price/earnings ratio AND returns on equity for each company for the last three quarters
for index, row in nasdaq_df.iterrows():
    ticker = row['Company']
    quarterly_ratio_url = f'https://financialmodelingprep.com/api/v3/ratios/{ticker}?period=quarter&limit=3&apikey={api_key}' 
    response = requests.get(quarterly_ratio_url).json()

    try:
        nasdaq_df.loc[index, 'Q3 P/E'] = response[0]['priceEarningsRatio']
        nasdaq_df.loc[index, 'Q2 P/E'] = response[1]['priceEarningsRatio']
        nasdaq_df.loc[index, 'Q1 P/E'] = response[2]['priceEarningsRatio']
        nasdaq_df.loc[index, 'Q3 ROE'] = response[0]['returnOnEquity']
        nasdaq_df.loc[index, 'Q2 ROE'] = response[1]['returnOnEquity']
        nasdaq_df.loc[index, 'Q1 ROE'] = response[2]['returnOnEquity']
        nasdaq_df.loc[index, 'Q3 Date'] = response[0]['date']
        nasdaq_df.loc[index, 'Q2 Date'] = response[1]['date']
        nasdaq_df.loc[index, 'Q1 Date'] = response[2]['date']
    except IndexError:
        print('Missing NASDAQ information')

    
```

```python
nasdaq_df.head()
```

```python
nasdaq_df.to_csv('output/nasdaq_financial_ratios.csv')
```

# NASDAQ Historical Prices

```python
nasdaq_url = 'https://financialmodelingprep.com/api/v3/nasdaq_constituent?apikey=' + api_key
print(nasdaq_url)
```

```python
response = requests.get(nasdaq_url).json()
print(json.dumps(response, indent=4, sort_keys=True))
```

```python
# List of Companies
nasdaqcompanies = []
nasdaqsymbols = []

for company in response:
    nasdaqcompanies.append(company['name'])
    nasdaqsymbols.append(company['symbol'])
```

```python
data = {'Company': nasdaqsymbols,
        'Q1 Date': ['' for i in range(len(nasdaqcompanies))],
        'Q1 Price': ['' for i in range(len(nasdaqsymbols))], 
        'Q2 Date': ['' for i in range(len(nasdaqsymbols))],
        'Q2 Price': ['' for i in range(len(nasdaqsymbols))], 
        'Q3 Date': ['' for i in range(len(nasdaqsymbols))],
        'Q3 Price': ['' for i in range(len(nasdaqsymbols))],  
       }

nasdaq_df = pd.DataFrame(data=data)
nasdaq_df.head()
```

```python
# Historical Data Collection

for index, row in nasdaq_df.iterrows():
        
    companysymbol = row["Company"]
    date_url = f'https://financialmodelingprep.com/api/v3/ratios/{companysymbol}?period=quarter&limit=3&apikey={api_key}'
    response = requests.get(date_url).json()    
    
    try:  
        
        nasdaq_df.loc[index, "Q1 Date"] = response[2]["date"]
        nasdaq_df.loc[index, "Q2 Date"] = response[1]["date"]
        nasdaq_df.loc[index, "Q3 Date"] = response[0]["date"]
        
    except IndexError:
        print(f"N/A")
        
   

    
print(nasdaq_df)
```

```python
nasdaq_df.head()    
```

```python
base_url = "https://financialmodelingprep.com/api/v3/historical-price-full/"
for index, row in nasdaq_df.iterrows():
    
    companysymbol = row["Company"]
    
    # Quarter 1 Price Response
    q1_date = row["Q1 Date"]
    q1_price_url = f'{base_url}{companysymbol}?from={q1_date}&to={q1_date}&apikey={api_key}'
    response = requests.get(q1_price_url).json() 
    
    
    try:  
        nasdaq_df.loc[index, "Q1 Price"] = response['historical'][0]['close']
        
    except (IndexError, KeyError):
        print(f"N/A")
    
    
    #Quarter 2 Price Response2
    q2_date = row["Q2 Date"]
    q2_price_url = f'{base_url}{companysymbol}?from={q2_date}&to={q2_date}&apikey={api_key}'
    response2 = requests.get(q2_price_url).json() 
    
    
    try:  
        nasdaq_df.loc[index, "Q2 Price"] = response2['historical'][0]['close']
    
    except (IndexError, KeyError):        
            print(f"N/A")
        
    
    #Quarter 3 Price Response3 
    q3_date = row["Q3 Date"]
    q3_price_url = f'{base_url}{companysymbol}?from={q3_date}&to={q3_date}&apikey={api_key}'
    response3 = requests.get(q3_price_url).json()
    
     
    try:  
        nasdaq_df.loc[index, "Q3 Price"] = response3['historical'][0]['close']
    
    except (IndexError, KeyError):
        print(f"N/A")
    
    
```

```python
nasdaq_df.head(10)
```

```python
nasdaq_df.to_csv('output/nasdaq_historical_prices.csv')
```

```python
data1 = {'Company': snp_company_name, 
         'Symbol': snp_company_symbol,
        'Q3 Date': ['' for i in range(len(snp_company_name))], 
        'Q3 Price': ['' for i in range(len(snp_company_name))], 
        'Q2 Date': ['' for i in range(len(snp_company_name))], 
        'Q2 Price': ['' for i in range(len(snp_company_name))], 
        'Q1 Date': ['' for i in range(len(snp_company_name))], 
        'Q1 Price': ['' for i in range(len(snp_company_name))] 
       }
snp_df = pd.DataFrame(data=data1)
snp_df.head(-1)
```

```python
for index, row in snp_df.iterrows():
        
    symbol = row["Symbol"]
    date_url = f'https://financialmodelingprep.com/api/v3/ratios/{symbol}?period=quarter&limit=3&apikey={api_key}'
    #price_url = f'https://financialmodelingprep.com/api/v3/historical-price-full/{symbol}?from={date}&to={date}&apikey={api_key}'
    response = requests.get(date_url).json()
    
    
    try:  
        
        snp_df.loc[index, "Q3 Date"] = response[0]["date"]
        snp_df.loc[index, "Q2 Date"] = response[1]["date"]
        snp_df.loc[index, "Q1 Date"] = response[2]["date"]
        
    except IndexError:
        print("Missing information")
```

```python
#get price 
for index, row in snp_df.iterrows():
    
    symbol = row["Symbol"]
    
    # Price Response for Q2 2020 Date
    date1 = row["Q3 Date"]
    date1_price_url = f'https://financialmodelingprep.com/api/v3/historical-price-full/{symbol}?from={date1}&to={date1}&apikey={api_key}'
    date1_response = requests.get(date1_price_url).json() 
    
    
    try:  
        snp_df.loc[index, "Q3 Price"] = date1_response['historical'][0]['close']
        
    except (IndexError, KeyError):
        print(f" - ")
    
    
    #Price Response for Q1 2020 Date
    date2 = row["Q2 Date"]
    date2_price_url = f'https://financialmodelingprep.com/api/v3/historical-price-full/{symbol}?from={date2}&to={date2}&apikey={api_key}'
    date2_response = requests.get(date2_price_url).json() 
    
    
    try:  
        snp_df.loc[index, "Q2 Price"] = date2_response['historical'][0]['close']
        
    except (IndexError, KeyError):
        print(f" - ")

            
    
    #Price Response for Q4 2019 Date
    date3 = row["Q1 Date"]
    date3_price_url = f'https://financialmodelingprep.com/api/v3/historical-price-full/{symbol}?from={date3}&to={date3}&apikey={api_key}'
    date3_response = requests.get(date3_price_url).json()
    
     
    try:  
        snp_df.loc[index, "Q1 Price"] = date3_response['historical'][0]['close']
    
    except (IndexError, KeyError):
        print(f" - ")
```
