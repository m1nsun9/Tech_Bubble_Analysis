```python
import json
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from config_sk import api_key
```

```python
#call snp 500 company name list
snp_url=f"https://pkgstore.datahub.io/core/s-and-p-500-companies/constituents_json/data/8fd832966a715a70cb9cf3f723498e3b/constituents_json.json"
print(requests.get(snp_url))
```

```python
snp_response = requests.get(snp_url).json()
#print(json.dumps(snp_response, indent=4, sort_keys=True))
```

```python
#-------------------------------------------------------------------------------------------------------
```

```python
#call historical price api
url=f"https://financialmodelingprep.com/api/v3/historical-price-full/%5EGSPC?apikey={api_key}"
print(requests.get(url))
```

```python
price_response = requests.get(url).json()
#print(json.dumps(price_response, indent=4, sort_keys=True))
```

```python
#-------------------------------------------------------------------------------------------------------
```

```python
# collect a list of the S&P 500 constituents
snp_company_name = []
snp_company_symbol = []


for company in snp_response:
    snp_company_name.append(company['Name'])
    snp_company_symbol.append(company['Symbol'])
```

```python
data1 = {'Name': snp_company_name, 
         'Symbol': snp_company_symbol,
        'Q3 Date': ['' for i in range(len(snp_company_name))], 
        'Q3 Price': ['' for i in range(len(snp_company_name))], 
        'Q2 Date': ['' for i in range(len(snp_company_name))], 
        'Q2 Price': ['' for i in range(len(snp_company_name))], 
        'Q1 Date': ['' for i in range(len(snp_company_name))], 
        'Q1 Price': ['' for i in range(len(snp_company_name))] 
       }
snp_df = pd.DataFrame(data=data1)
snp_df.head(-1)
```

```python
for index, row in snp_df.iterrows():
        
    symbol = row["Symbol"]
    date_url = f'https://financialmodelingprep.com/api/v3/ratios/{symbol}?period=quarter&limit=3&apikey={api_key}'
    #price_url = f'https://financialmodelingprep.com/api/v3/historical-price-full/{symbol}?from={date}&to={date}&apikey={api_key}'
    response = requests.get(date_url).json()
    
    
    try:  
        
        snp_df.loc[index, "Q3 Date"] = response[0]["date"]
        snp_df.loc[index, "Q2 Date"] = response[1]["date"]
        snp_df.loc[index, "Q1 Date"] = response[2]["date"]
        
    except IndexError:
        print("Missing information")
```

```python
#get price 
for index, row in snp_df.iterrows():
    
    symbol = row["Symbol"]
    
    # Price Response for Q2 2020 Date
    date1 = row["Q3 Date"]
    date1_price_url = f'https://financialmodelingprep.com/api/v3/historical-price-full/{symbol}?from={date1}&to={date1}&apikey={api_key}'
    date1_response = requests.get(date1_price_url).json() 
    
    
    try:  
        snp_df.loc[index, "Q3 Price"] = date1_response['historical'][0]['close']
        
    except (IndexError, KeyError):
        print(f" - ")
    
    
    #Price Response for Q1 2020 Date
    date2 = row["Q2 Date"]
    date2_price_url = f'https://financialmodelingprep.com/api/v3/historical-price-full/{symbol}?from={date2}&to={date2}&apikey={api_key}'
    date2_response = requests.get(date2_price_url).json() 
    
    
    try:  
        snp_df.loc[index, "Q2 Price"] = date2_response['historical'][0]['close']
        
    except (IndexError, KeyError):
        print(f" - ")

            
    
    #Price Response for Q4 2019 Date
    date3 = row["Q1 Date"]
    date3_price_url = f'https://financialmodelingprep.com/api/v3/historical-price-full/{symbol}?from={date3}&to={date3}&apikey={api_key}'
    date3_response = requests.get(date3_price_url).json()
    
     
    try:  
        snp_df.loc[index, "Q1 Price"] = date3_response['historical'][0]['close']
    
    except (IndexError, KeyError):
        print(f" - ")
```
