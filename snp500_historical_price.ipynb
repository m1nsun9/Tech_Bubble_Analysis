```python
# collect a list of the S&P 500 constituents
snp_company_name = []
snp_company_symbol = []


for company in snp_response:
    snp_company_name.append(company['name'])
    snp_company_symbol.append(company['symbol'])
```

```python
data1 = {'Company': snp_company_name, 
         'Symbol': snp_company_symbol,
        '2020-06-27': ['' for i in range(len(snp_company_name))], 
        '2020-03-28': ['' for i in range(len(snp_company_name))], 
        '2019-12-28': ['' for i in range(len(snp_company_name))] 
       }
snp_df = pd.DataFrame(data=data1)
snp_df.head(-1)
```

```python
data1 = {'Company': snp_company_name, 
         'Symbol': snp_company_symbol,
        'Q2 2020 Date': ['' for i in range(len(snp_company_name))], 
        'Q2 2020 Price': ['' for i in range(len(snp_company_name))], 
        'Q1 2020 Date': ['' for i in range(len(snp_company_name))], 
        'Q1 2020 Price': ['' for i in range(len(snp_company_name))], 
        'Q4 2019 Date': ['' for i in range(len(snp_company_name))], 
        'Q4 2019 Price': ['' for i in range(len(snp_company_name))] 
       }
snp_df = pd.DataFrame(data=data1)
snp_df.head(-1)
```

```python
###ignore this code########################*************************

for index, row in snp_df.iterrows():
    symbol = row['Symbol']

    price_url = f'https://financialmodelingprep.com/api/v3/historical-price-full/{symbol}?serietype=line&apikey={api_key}' 
    #quarter_date_url=f'https://financialmodelingprep.com/api/v3/ratios/{symbol}?period=quarter&limit=3&apikey=1e66f91abd67cd77b6d1a6265607e908'
    response = requests.get(price_url).json()  
    
    try:
        snp_df.loc[index, 'Q2 2020 Price'] = response['historical'][0]['close']
        snp_df.loc[index, 'Q1 2020 Price'] = response['historical'][1]['close']
        snp_df.loc[index, 'Q4 2019 Price'] = response['historical'][2]['close']
    
    except IndexError:
        print('Missing information')
```

```python
for index, row in snp_df.iterrows():
        
    symbol = row["Symbol"]
    date_url = f'https://financialmodelingprep.com/api/v3/ratios/{symbol}?period=quarter&limit=3&apikey={api_key}'
    #price_url = f'https://financialmodelingprep.com/api/v3/historical-price-full/{symbol}?from={date}&to={date}&apikey={api_key}'
    response = requests.get(date_url).json()
    
    
    try:  
        
        snp_df.loc[index, "Q2 2020 Date"] = response[0]["date"]
        snp_df.loc[index, "Q1 2020 Date"] = response[1]["date"]
        snp_df.loc[index, "Q4 2019 Date"] = response[2]["date"]
        
    except IndexError:
        print("Missing information")
```
